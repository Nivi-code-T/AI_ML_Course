## 1. An end to end scikit learn workflow

```python
#1.Get the data Ready
import pandas as pd
import numpy as np
doc1=pd.read_csv("data/Doc1.csv")
doc1
```

```python
#Cretae X (features matrix)
X=doc1.drop("target",axis=1)

#Create Y label
Y=doc1["target"]
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# 2.Choosing the right model and hyperparameters
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier()

#We'll keep the default hyperparameter
clf.get_params()
```

```python
#3. Fit the model to the training data
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test=train_test_split(X, Y, test_size=0.2)

```

```python
clf.fit(X_train,y_train);
```

```python
X_train
```

```python
#Make a prediction
y_label=clf.predict(np.array([0,1,2,3]))
```

```python
y_preds=clf.predict(X_test)
y_preds
```

```python
y_test
```

```python
#4.Evaluate the model on the training data and the test data
clf.score(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
print(classification_report(y_test,y_preds))
```

```python
confusion_matrix(y_test,y_preds)
```

```python
accuracy_score(y_test,y_preds)
```

```python
#5. Improve model
#Try different amount of n estimaters
np.random.seed(42)
for i in range(10,100,10):
    print (f"Trying model with {i} estomators")
    clf=RandomForestClassifier(n_estimators=i).fit(X_test,y_test)
    print(f"Model Accuracy on test set: {clf.score(X_test,y_test)*100 :.2f}%")
    print("")
    
    
```

```python
#6. Save a model and load it
import pickle
pickle.dump(clf,open("Random_first_model_1.pkl","wb"))
            
```

```python
loaded_model=pickle.load(open("Random_first_model_1.pkl","rb"))
loaded_model.score(X_test,y_test)
```
