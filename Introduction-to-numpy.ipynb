```python
sales_amount=np.array([[2,3,4],[3,4,5],[3,5,8],[8,9,10],[11,12,13]])
sales_amount
```

```python
weekely_sales=pd.DataFrame(sales_amount,
                               index=["mon","tue","wed","thu","fri"],columns=["almond Butter","peanut Butter","cashew Butter"])
weekely_sales                               
```

```python
prices=np.array([10,8,12])
prices
prices.shape
```

```python
Butter_prices=pd.DataFrame(prices.reshape(1,3),index=["Price"],columns=["almond Butter","peanut Butter","cashew Butter"])
Butter_prices
```

```python
total_sales=prices.dot(sales_amount.T)
total_sales
```

```python
weekely_sales
```

```python
total_sales
```

```python
weekely_sales.shape,Butter_prices.shape
```

```python
weekely_sales.T.shape
```

```python
daily_sales=Butter_prices.dot(weekely_sales.T)
daily_sales
```

```python
weekely_sales
```

```python
weekely_sales["Total ($)"]=daily_sales.T
weekely_sales
```
