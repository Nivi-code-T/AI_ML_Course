```python
#step 1: import matplotlib and getting it ready to use in jupyternotebook
%matplotlib inline
import matplotlib.pyplot as plt

#step 2: Prepare some data
x=[1,2,3,4]
y=[10,20,30,40]

#Step 3: Setup plot
fig, ax=plt.subplots(figsize=(4,4))

#step 4:Plot a data
ax.plot(x,y)

#step 5:Customize plot
ax.set(title="Sample Plot",xlabel="x-axis",ylabel="y-axis")

#save and show the plot
fig.savefig("images/Sample-Plot.png")
```

```python
#make plot from dictionary
data={"A":10, "B":20,"C":40}
fig, ax=plt.subplots()
ax.bar(data.keys(),height=data.values())
```

```python
data={"A":10, "B":20,"C":40}
fig,ax=plt.subplots()
ax.barh(data.keys(),data.values())

```

```python
#make some data for histogram
x=np.random.randn(1000)
fig,ax=plt.subplots()
ax.hist(x)
```

##Two Options for subplots

```python
#Subplots Option 1
fig, ((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,
                                       ncols=2,
                                       figsize=(5,5))
ax1.plot(data.keys(),data.values())
ax2.scatter(x,np.exp(x))
ax3.scatter(x,np.sin(x))
ax4.hist(x)
```

```python
#Subplots Option 2
fig, ax=plt.subplots(nrows=2,ncols=2,figsize=(5,5))
ax[0,0].plot(data.keys(),data.values())
ax[0,1].scatter(x,np.exp(x))
ax[1,0].scatter(x,np.sin(x))
ax[1,1].hist(x)
```
